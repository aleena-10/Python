```python
import random

runs = 0
airuns = 0

def firstbatting():
    global runs, airuns  
    print('you are going to bat first')
    while True:
        me = int(input('bat:'))
        if me not in [1, 2, 3, 4, 5, 6]:
            print('invalid input')
        else:
            ai = random.randint(1, 6)
            print('ai ball:', ai)
            runs = runs + me
            print('your current score is:', runs)
            if me == ai:
                print('ai took your wicket')
                print('ai is going to bat now')
                print('ai needs', (runs + 1), 'runs to win the match')
                while True:
                    me = int(input('ball:'))
                    if me not in [1, 2, 3, 4, 5, 6]:
                        print('invalid input')
                    else:
                        ai = random.randint(1, 6)
                        print('ai bat:', ai)
                        airuns = airuns + me
                        print('ai current score is:', airuns)
                        if airuns > runs:
                            print('good try but ai won the match...better luck next time')
                            return  # Added to exit the function
                        elif me == ai:
                            print('you are successful in taking the wicket of ai')
                            print('you won the match by ', (runs - airuns), 'runs')
                            return  # Added to exit the function

def firstballing():
    global runs, airuns  # Added to access the global variables
    print('you are going to ball first')
    while True:
        me = int(input('ball:'))
        if me not in [1, 2, 3, 4, 5, 6]:
            print('invalid input')
        else:
            ai = random.randint(1, 6)
            print('ai bat:', ai)
            airuns = airuns + ai
            print('ai current score is:', airuns)
            if me == ai:
                print('you took ai wicket')
                print('you are going to bat now')
                print('you need', (airuns + 1), 'runs to win the match')
                while True:
                    me = int(input('bat:'))
                    if me not in [1, 2, 3, 4, 5, 6]:
                        print('invalid input')
                    else:
                        ai = random.randint(1, 6)
                        print('ai ball:', ai)
                        runs = runs + me
                        print('your current score is:', runs)
                        if runs > airuns:
                            print('you won the match')
                            return  # Added to exit the function
                        elif me == ai:
                            print('ai successfully took your wicket')
                            print('ai won the match by ', (airuns - runs), 'runs')
                            return  # Added to exit the function

name = input('PLEASE ENTER YOUR NAME')
toss = input('enter odd or even')
me = int(input('enter any number between 1 and 10:'))

if me > 6 or me < 1:
    print('invalid input')
else:
    ai = random.randint(1, 6)
    print('input of ai', ai)
    if toss.upper() == 'EVEN':
        if (me + ai) % 2 == 0:
            print('you won the toss, choose to bat or ball')
            tossinput = input('bat or ball:')
            if tossinput == 'bat':
                firstbatting()
            elif tossinput == 'ball':
                firstballing()
        else:
            print('ai wins the toss')
            aitoss = random.choice(['bat', 'ball'])
            if aitoss == 'bat':
                print('ai is going to bat first')
                firstballing()
            elif aitoss == 'ball':
                print('ai is going to ball first')
                firstballing()
    elif toss.upper() == 'ODD':
        if (me + ai) % 2 != 0:
            print('you won the toss, choose to bat or ball')
            tossinput = input('bat or ball:')
            if tossinput == 'bat':
                firstbatting()
            elif tossinput == 'ball':
                firstballing()
        else:
            print('invalid input')
    else:
        print('ai wins the toss')
        aitoss = random.choice(['bat', 'ball'])
        if aitoss == 'bat':
            print('ai is going to bat first')
            firstballing()
        elif aitoss == 'ball':
            print('ai is going to ball first')
            firstbatting()






```
